```python
#Dataset - https://www.kaggle.com/datasets/eshummalik/disney/data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('disney_plus_shows.csv')
df.head()
```

```python
df.describe(include = 'all')
```

```python
df.info()
```

```python
df['runtime'] = df['runtime'].str.replace(' min', '', regex=False)
df['runtime'] = df['runtime'].str.replace(' h', '', regex=False)
df['runtime'] = pd.to_numeric(df['runtime'])
```

```python
print(df.isnull().sum())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.duplicated().sum()
```

```python
df.shape
```

```python
(df.isnull().sum() / len(df)) * 100
```

```python
df.pop('metascore')
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df['imdb_id'].isna().sum()
```

```python
df.dropna(subset = ['imdb_id'],inplace = True)
```

```python
df.isnull().sum()
```

```python
df['imdb_votes'] = df['imdb_votes'].astype(str).str.replace(',','', regex = True).replace('nan',np.nan).astype(float)
df['plot'].fillna('No description available', inplace = True)
df['rated'].fillna('Not rated', inplace = True)
df['released_at'].fillna(df['released_at'].mode()[0], inplace = True)
df['runtime'].fillna(df['runtime'].median(), inplace = True)
df['genre'].fillna('Not available', inplace = True)
df['director'].fillna('Unknown', inplace = True)
df['writer'].fillna('Unknown', inplace = True)
df['actors'].fillna('Unknown', inplace = True)
df['language'].fillna('Unknown', inplace = True)
df['country'].fillna('Unknown', inplace = True)
df['awards'].fillna('Unknown', inplace = True)
df['imdb_rating'].fillna(df['imdb_rating'].mean(), inplace = True)
df['imdb_votes'].fillna(df['imdb_votes'].median(), inplace = True)
```

```python
df.isnull().sum()
```

```python
#How many rows and columns are in the dataset?
df.shape
```

```python
df['type']
```

```python
#How many movies vs TV shows are there? (type column)
df.groupby(['type']).size().reset_index(name = 'Total Number')
```

```python
#Which year has the highest number of releases on Disney+?
df.groupby('year')['imdb_id'].count().reset_index(name = 'Total Count').sort_values(by = 'Total Count', ascending = False).head(1)
```

```python
#What are the top 10 most common genres?
df.groupby('genre')['imdb_id'].count().reset_index(name ='Count').sort_values(by = 'Count', ascending = False).head(10)
```

```python
#What is the average IMDb rating of movies vs TV shows?
round(df.groupby('type')['imdb_rating'].mean().reset_index(name = 'Average'),2)
```

```python
#Which movie/TV show has the longest runtime?
df.groupby('title')['runtime'].max().reset_index(name = 'Duration').sort_values(by  = 'Duration', ascending = False).head(1)
```

```python
#Which country produces the most Disney+ content?
df.groupby('country')['imdb_id'].count().reset_index(name='Count').sort_values(by='Count',ascending=False).head(1)
```

```python
df.columns
```

```python
#Which directors have the most content on Disney+?
df.loc[df.director != 'Unknown'].groupby('director')['imdb_id'].size().reset_index(name='Count').sort_values(by='Count', ascending=False).head(1)
```

```python
#What are the top 10 most common actors in Disney+ content?
df.loc[df.actors !='Unknown'].groupby('actors')['imdb_id'].size().reset_index(name='Count').sort_values(by='Count', ascending=False).head(10)
```

```python
#What is the trend of releases over time (per year)?
groupedDF = df.groupby('year')['imdb_id'].count().reset_index(name='Count')
groupedDF['Count'].mean()
```

```python
#Bar chart: count of movies vs TV shows.
sns.countplot(data=df,x='type')
plt.show()
```

```python
#Compare IMDb ratings of movies vs TV shows with boxplots.
sns.boxplot(data=df, x='type', y='imdb_rating')
plt.title('IMBD Ratings')
plt.xlabel('Type')
plt.ylabel('Ratings')
plt.tight_layout()
plt.show()
```

```python
#Find correlation between runtime, IMDb rating, and IMDb votes.
corr_matrix = df[['runtime','imdb_rating','imdb_votes']].corr()
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.title('Correlation between Runtime, IMDb Rating, and IMDb Votes')
plt.show()
```

```python
#Do movies with awards tend to have higher IMDb ratings?
df['has_awards'] = df['awards'].apply(lambda x:'yes' if x!='Unknown' else 'No')
df.groupby('has_awards')['imdb_rating'].mean()
sns.boxplot(data=df, x='has_awards', y='imdb_rating')
plt.show()
```

```python
#What is the distribution of IMDb votes
sns.histplot(data=df, x='imdb_votes',bins=10)
```

```python
#Find the top 5 hidden gems (high rating but low votes).
df[['title','imdb_rating','imdb_votes']].sort_values(by=['imdb_rating','imdb_votes'], ascending=[False,True])
```

```python
#Trend of releases by country (lineplot for top 5 countries).
top_countries = df.loc[df.country!='Unknown'].groupby('country')['imdb_id'].size().reset_index(name='Count').sort_values(by='Count',ascending=False).head()
sns.lineplot(data=top_countries,x='country',y='Count',marker='o')
```

```python
#Are longer movies better rated? (scatterplot runtime vs rating).
plt.scatter(df['imdb_rating'],df['runtime'],alpha=0.5,c='red')
plt.xlabel('Rating')
plt.ylabel('Runtime')
plt.title('Scatter Plot')
plt.show()
```
